### 1 Nearest Neighbor

1-Nearest Neighbor is a case of k-Nearest Neighbors where k = 1

The new data point can be classified into Class 0 or Class 1 by calculating the Euclidean distance to the existing data points and assigning the new point to the class of the nearest neighbor, as this is a 1-Nearest Neighbor classification method.

**Training Step**

Nearest Neighbor algorithms are referred to as lazy learning algorithms because they do not involve a formal training process; instead, they simply store the dataset for future reference.

**Prediction Step**

Given:
- $$ X^i, y^i $$

Predict:
- $$ f(x^q) $$

Algorithm:

```
closest_point = None
closest_distance = infinity

for i in range(n):
    current_distance = d(X^i, x^q)
    if current_distance < closest_distance:
        closest_distance = current_distance
        closest_point = X^i
        
return f(closest_point)
```

The most commonly used distance function is Euclidean Distance 


For two points with coordinates $ A = (x_1, x_2, \ldots, x_n) $ and $ B = (y_1, y_2, \ldots, y_n) $, the Euclidean distance $ d_{AB} $ is given by:

$d_{AB} = \sqrt{\sum_{i=1}^{n}(y_i - x_i)^2}$

### Big-O Notation and kNN Run Time complexity

For kNN The big O complexity is 

* For K neighbors for one training example
* Checking for n training examples in a data set
* Calculating Euclidean distance of M dimensions


So the time complexity is $O(k*n*m)$ usually k and m are small so k and m are ignored this means the expression is reduced to $O(n)$
