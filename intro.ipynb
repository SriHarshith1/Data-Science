### Machine Learning

Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed

For example a spam filter can be made using a machine learning algorithm . with inputs as emails and outputs as spam/ Not spam labeled in the outputs

**Applications of Machine Learning**

* Email Spam/ Non Spam Classification
* Face Recognition
* Weather Prediction

**Categories of Machine Learning**

**Supervised Learning:** Supervised learning is the subcategory of machine learning that focuses on learning a classification or regression model that is learning from labeled training data (i.e., inputs that also contain the desired outputs or targets; basically, “examples” of what we want to predict).


Unlike classification regresion deals with continous data points

**Unsupervised Learning :** that is concerned with unlabeled data. Common tasks in unsupervised learning are clustering analysis (assigning group memberships; Figure 6) and dimensionality reduction (compressing data onto a lower-dimensional subspace or manifold).


**Reinforcement Learning**
Reinforcement is the process of learning from rewards while performing a series of actions.In reinforcement learning, we do not tell the learner or agent (for example, a robot), which action to take but merely assign a reward to each action and/or the overall outcome. Instead of having “correct/false” labels for each step, the learner must discover or learn a behavior that maximizes the reward for a series of actions
