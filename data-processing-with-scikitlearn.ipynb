### Basic Data Handling

* It is better to convert string class labels into integer class labels as this might result in less errors

This can be done applying a function to the whole data frame

```python
x = df['Species'].unique()
```

```python
lst = x.tolist()
```

```python
def species(y):
    for i in lst:
        if y == i :
            return lst.index(i)
```

```python
df['Class'] = df['Species'].apply(species)
```
